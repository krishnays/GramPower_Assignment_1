```python
import pyodbc
import textwrap


# define the server and the database
server = 'KRISHNAPC' 
database = 'exp1'  

# Define the connection string
cnxn = pyodbc.connect(
    'DRIVER={ODBC Driver 17 for SQL Server}; \
     SERVER='+ server +'; \
     DATABASE='+ database +';\
     Trusted_Connection=yes;')


# Create the Cursor.
cursor = cnxn.cursor()
  

    
# grab all the rows from the table
cursor.execute('SELECT * FROM usern')
for row in cursor:
    print(row)
    
# close the cursor and connection  
cursor.close()
cnxn.close()



```
